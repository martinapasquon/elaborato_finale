## Dimensione della popolazione totale ed efficace

```python
bins=500
n_gen=32
n_bott=7
n_gen_0=115


fig= plt.figure(figsize=(20,10))
possibili, v= np.loadtxt("C:../es_new/CRC0441_pop_grafico_bello.txt",usecols=(1,2), delimiter='     ', unpack='true')

i=np.linspace(0, len(v), len(v))



i=np.linspace(0, len(v), len(v))

for k in range(n_bott+1):
    plt.axvline(x=(n_gen*k) -1 , color= 'orange', linewidth=6.0,linestyle='dotted')
plt.axvline(x=n_gen*n_bott +n_gen_0, color='green',linewidth=6.0,linestyle='dotted')
plt.plot(i, possibili, color="red", linestyle="solid",linewidth=6.0, label="dimensione popolazione efficace")
plt.plot(i, v,   color="black", linewidth=6.0, alpha=0.3,label="dimensione popolazione totale")

plt.xlim([0,350])
plt.yscale("log")

plt.xlabel("tempo nel passato dalla fine dell'esperimento [generazioni]")
plt.ylabel('cellule vive')
plt.grid(axis='y',linestyle='dotted', color='grey')
plt.show()

fig = plt.figure(figsize=(20, 10))
for k in range(n_bott+1):
    plt.axvline(x=(n_gen*k) -1 , color= 'orange' ,linewidth=6.0, linestyle='dotted')
plt.axvline(x=n_gen*n_bott +n_gen_0, color='green', linewidth=6.0,linestyle='dotted')
plt.plot(i, v, linestyle="solid", color="black", linewidth=6.0, label="dimensione popolazione totale")
plt.xlim([0,350])
plt.yscale("log")
plt.xlabel("tempo nel passato dalla fine dell'esperimento [generazioni]")
plt.ylabel('cellule vive')

plt.grid(axis='y',linestyle='dotted', color='grey')
plt.show()
```

## Distribuzione dei tempi di coalescenza
In questo grafico sono rappresentate le distribuzioni dei tempi di coalescenza $\tau_2$ e $\tau_3$ per la linea cellulare
CRC0441 con $N_{bott}$ = 7 bottlenecks. Entrambi gli istogrammi sono stati ottenuti
con 50 000 simulazioni.

```python

bins=70


fig = plt.figure(figsize=(10, 10))

rami_1, rami_2= np.loadtxt("C:../es_new/CRC0441_grafico.txt",usecols=(0, 1), delimiter='      ', unpack='true')

for k in range(n_bott+1):
    plt.axvline(x=(n_gen*k) -1 , color= 'orange', linewidth=3.0,linestyle='dotted')
plt.axvline(x=n_gen*n_bott +n_gen_0, color='green',linewidth=3.0,linestyle='dotted')
plt.hist(rami_1,bins,  density=True, facecolor='b', label="istogramma lunghezza rami tipo 1")



plt.title(  r"$\tau_2$")

plt.ylabel('probabilit')
plt.xlabel('generazioni')
plt.xlabel("tempo nel passato dalla fine dell'esperimento")
plt.yscale("log")
plt.grid(axis='y',linestyle='dotted', color='grey')
plt.xlim([0,350])
plt.ylim([10E-6,0.1])

fig = plt.figure(figsize=(10, 10))
for k in range(n_bott+1):
    plt.axvline(x=(n_gen*k) -1 , color= 'orange', linewidth=3.0,linestyle='dotted')
plt.axvline(x=n_gen*n_bott +n_gen_0, color='green',linewidth=3.0,linestyle='dotted')
plt.hist(rami_2,bins,  density=True, facecolor='b', label="istogramma lunghezza rami tipo 2")

plt.grid(axis='y',linestyle='dotted', color='grey')
plt.title( r"$\tau_3$")
plt.xlabel("tempo nel passato dalla fine dell'esperimento")
plt.yscale("log")
plt.xlim([0,350])
plt.ylim([10E-6,0.1])


```
