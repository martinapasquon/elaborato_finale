## La probabilità di avere coalescenza ottenuta dai dati è in accordo con la probabilità calcolata analiticamente
est effettuato con $N_0=2, 4, 8, 16, 10, 100$ cellule. $P_a=1$. Ogni grafico rappresenta il valor medio su 10000000 simulazioni.

```python
def test(n_gen, N_0):
    p=np.zeros(n_gen)
    a=(N_0*np.power(2, n_gen))
    for i in range(len(p)):
        p[i]=np.power(2, i)/(a-1)
    p[0]=0
    return p
```

```python
def percentuale(rami, n_gen, n_tr):
    conta=np.zeros(n_gen)
    for i in range(len(rami)):
        for j in range(n_gen):
            if (rami[i]>j and rami[i]<=(j+1)):
                conta[j]=conta[j]+1
                   # print(conta[j,k])
    
   
    conta=conta/n_tr
   # print(conta)   
    return conta
   
```

```python
bins=50
bins2=50
n_gen=15
N_0=8
n_tr=10000000

rami_100= np.loadtxt("C:../test/test_paper_100cell.txt",usecols=(0), delimiter=' ', unpack='true')
rami_40= np.loadtxt("C:../test/test_paper_40cell.txt",usecols=(0), delimiter=' ', unpack='true')
rami_16= np.loadtxt("C:../test/test_paper_16cell.txt",usecols=(0), delimiter=' ', unpack='true')
rami_8= np.loadtxt("C:../test/test_paper_8cell.txt",usecols=(0), delimiter=' ', unpack='true')
rami_4= np.loadtxt("C:../test/test_paper_4cell.txt",usecols=(0), delimiter=' ', unpack='true')
rami_2= np.loadtxt("C:../test/test_paper_2cell_1.txt",usecols=(0), delimiter=' ', unpack='true')


p_100=test(n_gen,100) 
p_40=test(n_gen,40) 
p_16=test(n_gen,16) 
p_8=test(n_gen,8) 
p_4=test(n_gen,4) 
p_2=test(n_gen,2) 
x=np.linspace(0, len(p_100), len(p_100))
conta_100 =percentuale(rami_100, n_gen, n_tr)
conta_tot100=len(rami_100)/(n_tr)
conta_40 =percentuale(rami_40, n_gen, n_tr)
conta_tot40=len(rami_40)/(n_tr)
conta_16 =percentuale(rami_16, n_gen, n_tr)
conta_tot16=len(rami_16)/(n_tr)
conta_8 =percentuale(rami_8, n_gen, n_tr)
conta_tot8=len(rami_8)/(n_tr)
conta_4 =percentuale(rami_4, n_gen, n_tr)
conta_tot4=len(rami_4)/(n_tr)
conta_2 =percentuale(rami_2, n_gen, n_tr)
conta_tot2=len(rami_2)/(n_tr)

pow_2=np.power(2, n_gen)

p_out_100=1-(pow_2-1)/((100*pow_2)-1)
p_out_40=1-(pow_2-1)/((40*pow_2)-1)
p_out_16=1-(pow_2-1)/((16*pow_2)-1)
p_out_8=1-(pow_2-1)/((8*pow_2)-1)
p_out_4=1-(pow_2-1)/((4*pow_2)-1)
p_out_2=1-(pow_2-1)/((2*pow_2)-1)


sns.set_palette("Paired", n_colors=13, desat=None, color_codes=False)
fig = plt.figure(figsize=(10, 10))
  
area = 1000



plt.scatter(x,p_100, s=area,  marker='o', color='y', alpha=0.3,  label="modello")
plt.scatter(x,conta_100,s=area,  marker='x', color='y',  label="simulazione a tempo inverso")
plt.scatter(x,p_40, s=area,  marker='o', color='purple',  alpha=0.3,   label="modello")
plt.scatter(x,conta_40,s=area,  marker='x',  color='purple', label="simulazione a tempo inverso")
plt.scatter(x,p_16, s=area,  marker='o',  color='orange', alpha=0.3,  label="modello")
plt.scatter(x,conta_16,s=area,  marker='x', color='orange', label="simulazione a tempo inverso")
plt.scatter(x,p_8, s=area,  marker='o',color='r', alpha=0.3,   label="modello")
plt.scatter(x,conta_8,s=area,  marker='x', color='r', label="simulazione a tempo inverso")
plt.scatter(x,p_4, s=area,  marker='o',color='g', alpha=0.3,   label="modello")
plt.scatter(x,conta_4,s=area,  marker='x',color='g',  label="simulazione a tempo inverso")
plt.scatter(x,p_2, s=area,  marker='o', color='b',alpha=0.3,   label="modello")
plt.scatter(x,conta_2,s=area,  marker='x', color='b', label="simulazione a tempo inverso")

plt.title("coalescenza ")
plt.grid(axis='y',linestyle='dotted', color='grey')
plt.xlabel("tempo nel passato dalla fine dell'esperimento")
plt.xlim(0.,16)
plt.ylabel('probabilità')
plt.ylim(10E-8,1.5)
plt.yscale("log")
sns.set_palette("Paired", n_colors=13, desat=None, color_codes=False)


plt.show()

```

Test effettuato con $P_a=0.5, 0.6, 0.7, 0.8, 0.9.$ Ogni grafico è stato ottenuto con 50000 simulazioni. $N_0=50 $ cellule, $N_{gen}=10 $ generazioni, $N_{cellule}=3 $ cellule. In ogni simuazione si sono presi i dati relativi alle utlime 5 generazioni.

```python
def probabilita(rate, n, delta):
    prob=np.zeros((5, 2))
    prob[0,0]=(0.5)*(0.5)
    prob[0,1]=2*(0.5)*(1-0.5)
    for i in range(0,4):
        prob[i+1,0]=(rate+delta*i)*(rate+delta*i)
        prob[i+1,1]=2*(rate+delta*i)*(1-rate-delta*i)
    return prob
```

```python
def probabilita_norm(rate, n, delta):
    prob=np.zeros((5, 3 ))
    prob[0,0]=0
    prob[0,1]=(0.5)
    prob[0,2]=(0.5)*(0.5)
    somma=prob[0,1]+prob[0,2]
    prob[0,1]=prob[0,1]/somma
    prob[0,2]=prob[0,2]/somma
    
    for i in range(0,4):
        prob[i+1,2]=(rate+delta*i)*(rate+delta*i)
        prob[i+1,1]=2*(rate+delta*i)*(1-rate-delta*i)
        somma=(prob[i+1,2]+prob[i+1,1])
        prob[i+1,2]=prob[i+1,2]/somma
        prob[i+1,1]=prob[i+1,1]/somma
        prob[i+1,0]=0
        
    return prob
```
